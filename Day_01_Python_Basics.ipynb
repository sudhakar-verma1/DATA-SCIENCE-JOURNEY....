** MINI PROJECT — Student Report Card**

```python
student_name = "Rahul Verma"
subject1 = 85
subject2 = 90
subject3 = 78
total = subject1 + subject2 + subject3
average = total / 3
percentage = (total / 300)*100

print ("=" * 35)
print (" STUDENT REPORT CARD")
print ("=" * 35)
print (f"Student Name : {student_name}")
print (f"Student 1 :{subject1}")
print (f"Subject 2 : {subject2}")
print (f"Student 3 : {subject3}")
print ("-"*35)
print (f"Total Marks : {total}/300")
print (f"Average : {average:.2f}")
print (f"Percentage :{percentage:.2f}%")

if percentage >=90:
  grade="A+"
if percentage >=80:
  grade = "A"
if percentage >=70:
  grade = "B"
else:
  grade = "C"
print(f"Grade :{grade}")
print("="*35)
```
